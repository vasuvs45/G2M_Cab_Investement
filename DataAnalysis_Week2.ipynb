#### Merged_Final_DataSet

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
final_data_set=pd.read_csv('final_data.csv')
final_data_set.columns
df_cab = pd.read_csv('Cab_Data.csv')
df_holiday = pd.read_csv('US Holiday Dates (2004-2021).csv')
#cab_df = pd.read_csv('YOURDIRECTORY/Cab_Data.csv')
final_data_set['Date of Travelling'] = pd.to_datetime(df_cab['Date of Travel'], origin='1899-12-30', unit='D')
#final_data_set['Date of Travel'] = final_data_set['Date of Travel'].dt.strftime('%-m/%-d/%Y')
final_data_set['Date of Travelling']
final_data_set['travel_date']=  pd.to_datetime(final_data_set['Date of Travelling'])
df_holiday['holiday_date'] = pd.to_datetime(df_holiday['Date'])


final_data_set['Number of Customers'] = final_data_set.groupby('travel_date')['Transaction ID'].transform('count')

merged_final_data_set = pd.merge(final_data_set, df_holiday['holiday_date'], left_on='travel_date', right_on='holiday_date', how='left') 



merged_final_data_set['is_holiday'] = merged_final_data_set['holiday_date'].notnull()

merged_final_data_set.columns

total_customers = merged_final_data_set.groupby('is_holiday')['Number of Customers'].sum()

total_customers

plt.bar(['Regular Days', 'Holidays'], total_customers)
plt.xlabel( 'average_customer')
plt.ylabel('Average Number of Customers')
plt.title('Average Number of Customers on Holiday vs Non-Holiday Dates')
plt.show()



regular_days_data = merged_final_data_set[merged_final_data_set['is_holiday'] == False]

holiday_data = merged_final_data_set[merged_final_data_set['is_holiday'] == True]

regular_days_customers = regular_days_data.groupby('Company')['Number of Customers'].sum()


holiday_customers = holiday_data.groupby('Company')['Number of Customers'].sum()

companies = ['Pink Cab', 'Yellow Cab'] 
x = range(len(companies))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x, regular_days_customers, width, label='Regular Days')
rects2 = ax.bar([i + width for i in x], holiday_customers, width, label='Holidays')

ax.set_xlabel('Company')
ax.set_ylabel('Total Number of Customers')
ax.set_title('Comparison of Total Customers by Company on Regular Days vs Holidays')
ax.set_xticks([i + width/2 for i in x])
ax.set_xticklabels(companies)
ax.legend()

plt.show()
merged_final_data_set.columns
#merged_final_data_set.drop('Date of Travel',axis = 1,inplace = True)
merged_final_data_set.drop('Date of Travelling',axis = 1,inplace = True)
merged_final_data_set.to_csv('merged_final_data.csv')
```

As we can see in the above graphs, ,the sum of people travelling on Regular days is much greater than the sum of people travelling on Holidays. Hence it can be said that most people might use cab service to go to offices rather than travelling to holiday destinations. This also implies that either people don't like to travel on holidays and rather enjoy their holidays at home or they don't like to travel to holiday destinations through cabs. They might either use their own vehicle or travel via other modes of travel such as Flight, train etc.
